```python
import os
from collections import defaultdict

# Set path to labels directory (should contain .txt files for each image)
labels_dir =  r'C:\Users\Sujan.S\Videos\Smartan-Hackathon-2-SUJAN-S-22PD35-\SmartanAI-YOLO-Custom-Detection\dataset\train\labels'  # change this!

# Number of classes and class names (edit if different)
class_names = [
    'borosil_bottle', 'daisy_flower', 'hammer', 'hibiscus_flower',
    'plastic_bottle', 'rose_flower', 'screwdriver', 'spanner', 'tupperware_bottle'
]
num_classes = len(class_names)

# Initialize counters
label_count = defaultdict(int)
image_set = defaultdict(set)

# Process all label files
for label_file in os.listdir(labels_dir):
    if label_file.endswith(".txt"):
        filepath = os.path.join(labels_dir, label_file)
        with open(filepath, "r") as f:
            lines = f.readlines()
            for line in lines:
                class_id = int(line.strip().split()[0])
                if 0 <= class_id < num_classes:
                    label_count[class_id] += 1
                    image_set[class_id].add(label_file.replace('.txt', ''))

# Print summary
print("\n📊 Class-wise Summary:")
for i in range(num_classes):
    name = class_names[i]
    labels = label_count[i]
    images = len(image_set[i])
    print(f"{name:22} | Labels: {labels:<3} | Images: {images}")

# Total files
total_files = len([f for f in os.listdir(labels_dir) if f.endswith(".txt")])
print(f"\n🖼️ Total images: {total_files}")
```
